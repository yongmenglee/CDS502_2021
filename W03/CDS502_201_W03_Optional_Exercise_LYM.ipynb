## Tic Tac Toe

Develop a simple two-player Tic Tac Toe game in Python. The output of the Tic Tac
Toe game shall look something like this: 

```python
# Helper functions
def draw_grids():
    grids = ""
    for row in range(4):
        if row == 0:
            grids += " _ _ _ "
        else:
            grids += "|_|_|_|"
            
        if row != 3:
            grids += "\n"
    
    return grids

def get_index(row, col):
    index = (row) * 8 + (col*2) - 1
    
    return index

def is_valid_input(val):
    if val < 1 or val > 3:
        return False
    return True

def check_winner(grids, player1="X", player2="0"):
    winner = None
    
    if grids[get_index(1,1)] == grids[get_index(1,2)] == grids[get_index(1,3)]:
        winner = 1 if grids[get_index(1,1)] == player1 else 2 if grids[get_index(1,1)] == player2 else None
        
    elif grids[get_index(2,1)] == grids[get_index(2,2)] == grids[get_index(2,3)]:
        winner = 1 if grids[get_index(2,1)] == player1 else 2 if grids[get_index(2,1)] == player2 else None
        
    elif grids[get_index(3,1)] == grids[get_index(3,2)] == grids[get_index(3,3)]:
        winner = 1 if grids[get_index(3,1)] == player1 else 2 if grids[get_index(3,1)] == player2 else None
        
    elif grids[get_index(1,1)] == grids[get_index(2,1)] == grids[get_index(3,1)]:
        winner = 1 if grids[get_index(1,1)] == player1 else 2 if grids[get_index(1,1)] == player2 else None
        
    elif grids[get_index(1,2)] == grids[get_index(2,2)] == grids[get_index(3,2)]:
        winner = 1 if grids[get_index(1,2)] == player1 else 2 if grids[get_index(1,2)] == player2 else None
        
    elif grids[get_index(1,3)] == grids[get_index(3,2)] == grids[get_index(3,3)]:
        winner = 1 if grids[get_index(1,3)] == player1 else 2 if grids[get_index(1,3)] == player2 else None
        
    elif grids[get_index(1,1)] == grids[get_index(2,2)] == grids[get_index(3,3)]:
        winner = 1 if grids[get_index(1,1)] == player1 else 2 if grids[get_index(1,1)] == player2 else None
        
    elif grids[get_index(1,3)] == grids[get_index(2,2)] == grids[get_index(3,1)]:
        winner = 1 if grids[get_index(1,3)] == player1 else 2 if grids[get_index(1,3)] == player2 else None
    
    return winner
```

```python
# Play games
def start_game():
    current_player = 1
    
    grids = draw_grids()
    print(grids)
    winner = None
    
    round = 1
    
    while (winner is None) & (round <= 9) :
        print("Currently player: {}".format(current_player))
        
        row_num = int(input("Which row? (start with 1) "))
        col_num = int(input("Which column? (start with 1) "))
        
        if is_valid_input(row_num) & is_valid_input(col_num):
            
            index = get_index(row_num, col_num)
            
            if grids[index] != "_":
                continue

            mark = "X" if (current_player==1) else "0"
            grids = grids[:index] + mark + grids[index+1:]

            print(grids)
            
            winner = check_winner(grids, "X", "0")
            
            if winner is None:
                # Next player
                current_player = (current_player % 2) + 1
                round += 1
            else:
                print("Player {} wins".format(winner))
                
    if winner is None:
        print("Draw!")
```

```python
start_game()
```
