## Tic Tac Toe

Develop a simple two-player Tic Tac Toe game in Python.

**Note**: *The original exercise was completed within 1 hour (50 minute) as a programming challenge during lecture hours. This is the refactored version of the same Tic Tac Toe program.*

```python
# Helper functions
def draw_grids():
    grids = ""
    for row in range(4):
        if row == 0:
            grids += " _ _ _ "
        else:
            grids += "|_|_|_|"
            
        if row != 3:
            grids += "\n"
    
    return grids

def get_index(row, col): return (row*8) + (col*2) - 1
    
def is_valid_input(val): return False if ((val<1) or (val>3)) else True

def check_winner(grids, player1="X", player2="0"):
    # Assign winner
    def assign_winner(mark):
        return 1 if (mark==player1) else (2 if (mark==player2) else None)
    
    # Check horizontal/vertical
    for i in range(3):
        # - Horizontal: Rows
        grid_left = grids[get_index(i+1, 1)]
        if grid_left == grids[get_index(i+1, 2)] == grids[get_index(i+1, 3)]:
            return assign_winner(grid_left)
        
        # - Vertical: Columns
        grid_top = grids[get_index(1, i+1)]
        if grid_top == grids[get_index(2, i+1)] == grids[get_index(3, i+1)]:
            return assign_winner(grid_top)
    
    # Check diagonal
    diag_grid_center = grids[get_index(2,2)]
    for i in range(2):
        diag_grid_top = grids[get_index(1, (2*i)+1)]
        diag_grid_btm = grids[get_index(3, -(2*i)+3)]
        
        if diag_grid_center == diag_grid_top == diag_grid_btm:
            return assign_winner(diag_grid_center)
    
    return None

# Play games
def start_game():
    grids = draw_grids()
    winner = None
    current_player = 1
    round = 1
    show_player = True
    
    print(grids)
    
    while (winner is None) & (round <= 9):
        if show_player:
            print("Currently player: {}".format(current_player))
        
        row_num = int(input("Which row? (start with 1) "))
        col_num = int(input("Which column? (start with 1) "))
        
        if is_valid_input(row_num) & is_valid_input(col_num):
            show_player = True
            index = get_index(row_num, col_num)
            
            if grids[index] != "_":
                show_player = False
                continue

            mark = "X" if (current_player==1) else "0"
            grids = grids[:index] + mark + grids[index+1:]

            print(grids)
            
            winner = check_winner(grids, "X", "0")
            
            if winner is None:
                # Next player
                current_player = (current_player % 2) + 1
                round += 1
            else:
                print("Player {} wins!".format(winner))
                
        else:
            show_player = False
                
    if winner is None:
        print("Draw!")
```

```python
start_game()
```
